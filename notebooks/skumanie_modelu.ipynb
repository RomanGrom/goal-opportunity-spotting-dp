```python
# Importovanie knižníc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Načítanie dát z CSV
csv_path = "../.scratch/frame_predictions_zero.csv" 
data1 = pd.read_csv(csv_path)
```

```python
# Zobrazenie prvých riadkov
print("Prvých 5 riadkov dát:")
display(data1.head())
```

```python
# Základné štatistiky
print("\nZákladné štatistiky:")
display(data1.describe())
```

```python
# Počet jednotlivých akcií pre data1
action_counts_data1 = data1["Action"].value_counts().sort_index()

# Počet jednotlivých akcií pre data2
action_counts_data2 = data1["Action"].value_counts().sort_index()

# Nastavenie plochy grafov vedľa seba
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Graf pre data1
sns.barplot(ax=axes[0], x=action_counts_data1.index, y=action_counts_data1.values, palette="viridis")
axes[0].set_title("Rozdelenie predikovaných akcií - Data1", fontsize=16)
axes[0].set_xlabel("Akcia", fontsize=12)
axes[0].set_ylabel("Počet", fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

# Graf pre data2
sns.barplot(ax=axes[1], x=action_counts_data1.index, y=action_counts_data1.values, palette="viridis")
axes[1].set_title("Rozdelenie predikovaných akcií - Data2", fontsize=16)
axes[1].set_xlabel("Akcia", fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

# Zobrazenie grafov
plt.tight_layout()
plt.show()
```

```python
# Nastavenie plochy grafov vedľa seba
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Distribúcia hodnôt stavov pre data1
sns.histplot(data1["Value"], bins=30, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Distribúcia hodnôt stavov - Data1", fontsize=16)
axes[0].set_xlabel("Hodnota stavu", fontsize=12)
axes[0].set_ylabel("Počet", fontsize=12)

# Distribúcia hodnôt stavov pre data2
sns.histplot(data1["Value"], bins=30, kde=True, color="green", ax=axes[1])
axes[1].set_title("Distribúcia hodnôt stavov - Data2", fontsize=16)
axes[1].set_xlabel("Hodnota stavu", fontsize=12)
axes[1].set_ylabel("Počet", fontsize=12)

# Zobrazenie grafov
plt.tight_layout()
plt.show()
```

```python
# Analyzovanie akcií s najvyššou pravdepodobnosťou
action_prob_columns = [col for col in data1.columns if col.startswith("Action_Prob_")]
data1["Max_Prob"] = data1[action_prob_columns].max(axis=1)
data1["Predicted_Action"] = data1[action_prob_columns].idxmax(axis=1).apply(lambda x: int(x.split("_")[-1]))

```

```python
# Nastavenie plochy grafov vedľa seba
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Distribúcia najvyššej pravdepodobnosti pre data1
sns.histplot(data1["Max_Prob"], bins=30, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Distribúcia najvyšších pravdepodobností - Data1", fontsize=16)
axes[0].set_xlabel("Pravdepodobnosť", fontsize=12)
axes[0].set_ylabel("Počet", fontsize=12)

# Distribúcia najvyššej pravdepodobnosti pre data2

# Zobrazenie grafov
plt.tight_layout()
plt.show()
```

```python
# Priemerné hodnoty stavov pre jednotlivé akcie
avg_value_per_action1 = data1.groupby("Action")["Value"].mean()
avg_value_per_action2 = data1.groupby("Action")["Value"].mean()

# Zobrazenie vedľa seba
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Data1
sns.barplot(x=avg_value_per_action1.index, y=avg_value_per_action1.values, palette="Blues_d", ax=axes[0])
axes[0].set_title("Priemerná hodnota stavov podľa akcií - Data1", fontsize=16)
axes[0].set_xlabel("Akcia", fontsize=12)
axes[0].set_ylabel("Priemerná hodnota stavu", fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

# Data2
sns.barplot(x=avg_value_per_action2.index, y=avg_value_per_action2.values, palette="Greens_d", ax=axes[1])
axes[1].set_title("Priemerná hodnota stavov podľa akcií - Data2", fontsize=16)
axes[1].set_xlabel("Akcia", fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

# Nastavenie vzdialenosti medzi grafmi
plt.tight_layout()
plt.show()
```

```python
# Zobrazenie korelácií
plt.figure(figsize=(10, 8))
correlation_matrix = data1.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Korelačná matica medzi jednotlivými stĺpcami", fontsize=16)
plt.show()
```

```python
# Vizualizácia vývoja hodnoty v čase pre obe dáta
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Vývoj pre data1
axes[0].plot(data1.index, data1["Value"], color="blue", alpha=0.7, label="Hodnota stavu")
axes[0].set_title("Vývoj hodnôt v čase - Data1", fontsize=14)
axes[0].set_xlabel("Čas (index)", fontsize=12)
axes[0].set_ylabel("Hodnota stavu", fontsize=12)
axes[0].legend()
axes[0].grid(True, linestyle="--", alpha=0.6)



plt.tight_layout()
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

autocorrelation_plot(data1["Value"], ax=axes[0])
axes[0].set_title("Autokorelácia hodnôt stavov - Data1", fontsize=14)


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 8), sharey=True)

# Data1 - Boxplot hodnôt podľa akcie
sns.boxplot(data=data1, x="Action", y="Value", ax=axes[0], palette="viridis")
axes[0].set_title("Hodnoty stavov podľa akcie - Data1", fontsize=14)
axes[0].set_xlabel("Akcia", fontsize=12)
axes[0].set_ylabel("Hodnota stavu", fontsize=12)
axes[0].tick_params(axis='x', rotation=45)


plt.tight_layout()
plt.show()
```

```python
# Heatmapa distribúcie pravdepodobností
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.heatmap(data1.filter(like="Prob_").T, cmap="viridis", ax=axes[0])
axes[0].set_title("Distribúcia pravdepodobností akcií - Data1", fontsize=14)
axes[0].set_xlabel("Index predikcie", fontsize=12)
axes[0].set_ylabel("Akcia", fontsize=12)


plt.tight_layout()
plt.show()
```
