```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Načítanie dát z CSV
csv_path1 = "../.scratch/video_predictions_augmented.csv" 
df1 = pd.read_csv(csv_path1)

# Načítanie dát z CSV
csv_path2 = "../.scratch/video_predictions2.csv" 
df2 = pd.read_csv(csv_path2)
```

```python
print("Prvých 5 riadkov dát:")
display(df2.head())
```

```python
# Importy
import matplotlib.pyplot as plt
import seaborn as sns

# Nastavenie pre grafy
sns.set(style="whitegrid")

# Vytvoríme subplots pre df1 a df2
fig, axes = plt.subplots(1, 2, figsize=(20, 8), sharey=True)

# Dátový rámec 1 - df1
action_indexes_1 = df1['Action_Index'].unique()
for action in action_indexes_1:
    action_data = df1[df1['Action_Index'] == action]
    axes[0].plot(action_data['Frame_Index'], action_data['Value'], label=f"Akcia {action}", marker='o')
axes[0].set_title("Vývoj hodnôt - df1", fontsize=16)
axes[0].set_xlabel("Frame ID", fontsize=12)
axes[0].set_ylabel("Hodnota (Value)", fontsize=12)
axes[0].legend(title="Akcie", fontsize=10)

# Dátový rámec 2 - df2
action_indexes_2 = df2['Action_Index'].unique()
for action in action_indexes_2:
    action_data = df2[df2['Action_Index'] == action]
    axes[1].plot(action_data['Frame_Index'], action_data['Value'], label=f"Akcia {action}", marker='o')
axes[1].set_title("Vývoj hodnôt - df2", fontsize=16)
axes[1].set_xlabel("Frame ID", fontsize=12)
axes[1].legend(title="Akcie", fontsize=10)

# Zobrazenie grafov
plt.tight_layout()
plt.show()
```

```python
# Nastavenie pre grafy
sns.set(style="whitegrid")

# Získame zoznam unikátnych akcií
action_indexes = df2['Action_Index'].unique()

# Vytvoríme graf pre každú akciu
plt.figure(figsize=(12, 8))

for action in action_indexes:
    # Zoskupíme podľa 'action_index' a vezmeme len riadky pre danú akciu
    action_data = df2[df2['Action_Index'] == action]
    
    # Vytvoríme graf vývoja hodnôt pre túto akciu
    plt.plot(action_data['Frame_Index'], action_data['Value'], label=f"Akcia {action}", marker='o')

# Titulok a popisky osí
plt.title("Vývoj hodnôt pre jednotlivé akcie gólov", fontsize=16)
plt.xlabel("Frame ID", fontsize=12)
plt.ylabel("Hodnota (Value)", fontsize=12)

# Pridáme legendu
plt.legend(title="Akcie", fontsize=10)

# Zobrazenie grafu
plt.show()
```
