```python
# Importovanie potrebných knižníc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Akcie a ich popis
action_dict = {
    0: 'Idle (No-op)',
    1: 'Left',
    2: 'Top-Left',
    3: 'Top',
    4: 'Top-Right',
    5: 'Right',
    6: 'Bottom-Right',
    7: 'Bottom',
    8: 'Bottom-Left',
    9: 'Long Pass',
    10: 'High Pass',
    11: 'Short Pass',
    12: 'Shot',
    13: 'Sprint',
    14: 'Release Direction',
    15: 'Release Sprint',
    16: 'Sliding',
    17: 'Dribble',
    18: 'Release Dribble'
}
```

```python
# Načítanie CSV súboru do DataFrame
csv_path = "action_values_last2.csv"
data = pd.read_csv(csv_path)


# 1. Histogram rozdelenia akcií - sumár počtu výskytov jednotlivých akcií
action_counts = data['Action'].value_counts().sort_index()

# Uistíme sa, že 'Action' je celočíselný typ (pre správne zobrazenie na osi x)
action_counts = action_counts.sort_index()

# Zobrazenie histogramu akcií, kde x-ová os bude predstavovať čísla akcií
plt.figure(figsize=(12, 6))
plt.bar(action_counts.index.astype(int), action_counts.values, alpha=0.7, color='blue')
plt.title("Frequency of Actions")
plt.xlabel("Action")
plt.ylabel("Frequency")
plt.xticks(action_counts.index.astype(int))  # Zabezpečenie celočíselného zobrazenia na osi x
plt.grid(True)
plt.show()

# 2. Histogram rozdelenia hodnôt stavov (State Values)
plt.figure(figsize=(12, 6))
plt.hist(data['Value'], bins=30, alpha=0.7, color='green')
plt.title("Distribution of State Values")
plt.xlabel("State Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Načítanie CSV súboru do DataFrame
csv_path = "action_values_last3.csv"
data = pd.read_csv(csv_path)


# 1. Histogram rozdelenia akcií - sumár počtu výskytov jednotlivých akcií
action_counts = data['Action'].value_counts().sort_index()

# Zobrazenie histogramu akcií
plt.figure(figsize=(12, 6))
bars = plt.bar(action_counts.index.astype(int), action_counts.values, alpha=0.7, color='blue')
plt.title("Frequency of Actions\n(Hodnoty po 10 zápasoch a 15000 krokoch)")
plt.xlabel("Action")
plt.ylabel("Frequency")
plt.xticks(action_counts.index.astype(int))  # Zabezpečenie celočíselného zobrazenia na osi x
plt.grid(True)

# Vytvorenie legendy s číslami akcií namiesto farieb
action_labels = [f"Action {i}: {action_dict.get(i, 'Unknown')}" for i in action_counts.index]
plt.legend(bars, action_labels, title="Actions", loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()

# 2. Histogram rozdelenia hodnôt stavov (State Values)
plt.figure(figsize=(12, 6))
plt.hist(data['Value'], bins=30, alpha=0.7, color='green')
plt.title("Distribution of State Values\n(Hodnoty po 10 zápasoch a 15000 krokoch)")
plt.xlabel("State Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
